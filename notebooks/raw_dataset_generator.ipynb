```python
from scipy.interpolate import interp1d

def interpolate_rate(row, tau_target):
    """Interpolate r(tau_target) from rf_* columns in a given row."""
    rate_cols = [col for col in row.index if col.startswith('rf_') and pd.notna(row[col])]
    if len(rate_cols) < 2:
        return np.nan  # insufficient data

    taus = [float(col.replace('rf_', '')) for col in rate_cols]
    rates = [row[col] / 100 for col in rate_cols]

    f_interp = interp1d(taus, rates, kind='linear', fill_value='extrapolate')
    return float(f_interp(tau_target))

#interpolate_rate(df.iloc[], 1.0)  # Example for tau_target = 1.0 year
```

```python
def get_market_inputs(df, requested_date, maturity, maturity_to_tau):
    tau = maturity_to_tau[maturity]
    available_dates = pd.to_datetime(df.index)
    requested_dt = pd.to_datetime(requested_date)
    sample_date = available_dates[available_dates >= requested_dt].min().strftime('%Y-%m-%d')
    row = df.loc[sample_date]
    spot = row['last']
    q = row['dividend_yield'] / 100
    r = interpolate_rate(row, tau)
    return sample_date, tau, spot, q, r, row

get_market_inputs(df, '2024-10-01', '1Y', {'1Y': 1.0, '2Y': 2.0, '3Y': 3.0, '5Y': 5.0, '10Y': 10.0})
```

```python
from models.ssvi import extract_ssvi_surface, ssvi_iv, plot_ssvi_reconstruction

# maturity column parsing
iv_cols = [col for col in df.columns if col.startswith('iv_')]

def parse_iv_col(col):
   parts = col.split('_')
   if len(parts) != 3:
       return None
   _, strike, maturity = parts
   try:
       return col, float(strike), maturity
   except:
       return None

# maturity mapping (required for both heston & ssvi)
maturity_map = {}
for parsed in filter(None, map(parse_iv_col, iv_cols)):
   col, strike, maturity = parsed
   maturity_map.setdefault(maturity, []).append((col, strike))

# fixed maturity to tau mapping
maturity_to_tau = {
   '1m': 1/12, '2m': 2/12, '3m': 3/12, '6m': 0.5, '9m': 0.75,
   '1y': 1.0, '18m': 1.5, '2y': 2.0, '3y': 3.0, '4y': 4.0, '5y': 5.0
}

sample_date, tau, spot, q, r, row = get_market_inputs(df, '2024-03-16', '1y', maturity_to_tau)
print(f"sample date: {sample_date}, tau: {tau}, spot: {spot}, q: {q}, r: {r}")

ssvi_df = extract_ssvi_surface(df, maturity_map, maturity_to_tau)
plot_ssvi_reconstruction(
   df, ssvi_df, maturity_map, maturity_to_tau,
   interpolate_rate, "2024-03-18", "1y"
)
```

```python
import QuantLib as ql

def plot_heston_reconstruction(df, heston_df, maturity_map, maturity_to_tau, rate_func, date, maturity):
   sample_date, tau, spot, q, _, row = get_market_inputs(df, date, maturity, maturity_to_tau)

   pairs = maturity_map[maturity]
   ivs = np.array([row[col] for col, _ in pairs if pd.notna(row[col])])
   Ks = np.array([strike for col, strike in pairs if pd.notna(row[col])]) * spot / 100

   if (sample_date, maturity) not in heston_df.index:
       print(f"no calibrated heston params for ({sample_date}, {maturity})")
       return

   params = heston_df.loc[(sample_date, maturity)].to_dict()
   rate_ts = rate_func(row, tau)
   yield_ts = ql.YieldTermStructureHandle(
       ql.FlatForward(ql.Date.todaysDate(), q, ql.Actual365Fixed())
   )

   model_ivs = compute_heston_iv_curve(spot, rate_ts, yield_ts, tau, Ks, params)

   plt.figure(figsize=(10, 5))
   plt.plot(Ks / spot, ivs, 'o-', label='market iv')
   plt.plot(Ks / spot, model_ivs, 'x--', label='heston quantlib fit')
   plt.grid(True)
   plt.legend()
   plt.title(f"heston iv fit on {sample_date}, maturity={maturity}")
   plt.xlabel("moneyness (k/s)")
   plt.ylabel("implied volatility")
   plt.show()

plot_heston_reconstruction(
   df, heston_df, maturity_map, maturity_to_tau,
   rate_func=interpolate_rate_term_structure,
   date="2024-03-18",
   maturity="1y"
)
```

```python
def export_vol_tensor_csv(df, ssvi_df, heston_df, maturity_map, maturity_to_tau, out_path):
    records = []

    expected_strikes = np.round(np.arange(0.6, 1.51, 0.1), 2).tolist()

    for date, row in df.iterrows():
        S0 = row["last"]
        q = row["dividend_yield"] / 100

        # Interpolated rate curve
        rate_cols = [col for col in row.index if col.startswith("rf_")]
        sorted_rates = sorted(rate_cols, key=lambda x: float(x.replace("rf_", "")))
        curve = [row[col] / 100 for col in sorted_rates]

        for maturity, pairs in maturity_map.items():
            tau = maturity_to_tau.get(maturity)
            if tau is None or (date, maturity) not in ssvi_df.index or (date, maturity) not in heston_df.index:
                continue

            r = interpolate_rate(row, tau)
            if not np.isfinite(r): continue

            ssvi = ssvi_df.loc[(date, maturity)]
            heston = heston_df.loc[(date, maturity)]

            for col, strike in pairs:
                iv = row[col]
                if not np.isfinite(iv): continue

                rel_strike = round(strike / 100, 2)  # assumes strike was originally rel_strike * 100
                if rel_strike not in expected_strikes:
                    continue

                K = rel_strike * S0  # ensures K/S0 stays consistent across tau

                records.append({
                    "date": date,
                    "maturity": maturity,
                    "tau": tau,
                    "S0": S0,
                    "K": K,
                    "rel_strike": rel_strike,
                    "q": q,
                    "r": r,
                    "market_iv": iv,
                    **{f"rf_{i}": curve[i] for i in range(len(curve))},
                    **{f"ssvi_{k}": ssvi[k] for k in ssvi.index},
                    **{f"heston_{k}": heston[k] for k in heston.index}
                })

    pd.DataFrame(records).to_csv(out_path, index=False)
    print(f" Tensor dataset saved to {out_path}")

export_vol_tensor_csv(
    df, ssvi_df, heston_df, maturity_map, maturity_to_tau,
    out_path="data/vol_tensor_dataset_test.csv"
)
```
